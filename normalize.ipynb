```python

cate=file_data[file_data['cat']!='["uncategorised"]']
uncat=file_data[file_data['cat']=='["uncategorised"]']
```

```python
uncat_feat=[]
# you can change the length of the feature here
for index, row in uncat.iterrows():
    uncat_feat.append(((row['name'].lower()).split()[-3:],row['cat']))
cat_feat=[]
for index, row in cate.iterrows():
    cat_feat.append(((row['name'].lower()).split()[-3:],row['cat']))
```

```python
uncat_new=[]
for x in uncat_feat:
    max=0
    maxcat=[]
    for y in cat_feat:
        a=list(set(x[0]).intersection(set(y[0])))
        if len(a)>=max:
            max=len(a)
            maxcat=y[1]
    uncat_new.append((x[0],maxcat))  
```

```python
uncat_iter=0
for index,row in file_data.iterrows():
    if row['cat']=='["uncategorised"]':
        file_data.set_value(index, 'cat',uncat_new[uncat_iter][1])
        uncat_iter+=1
```

```python
newnan10=[]
for x in nan10['cat']:
    y=0
    div=0
    for index,row in file_data.iterrows():
        a=list(set(ast.literal_eval(x)).intersection(set(ast.literal_eval(row['cat']))))
#         print(x,set(ast.literal_eval(x)),a)
        if len(a)>=3:
            if row['purchases_Y15/M10']==row['purchases_Y15/M10']:
                y+=row['purchases_Y15/M10']
                div+=1
#             print(y)
    if div!=0:
#         print(int(y/div))
        newnan10.append(int(y/div))
    else:
#         print(y)
        newnan10.append(y)
```
